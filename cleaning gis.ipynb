```
import numpy as np
merged['Percent_No_Bus_Area'] = (1-(merged['No_Bus_Area'] / merged['Total Area'])) * 100
merged['Percent_No_Train_Area'] = (1-(merged['No_Train_Area'] / merged['Total Area'])) * 100
merged['PercentAreaNoGrocery'] = (1-(merged['Remaining Area no Grocery'] / merged['Total Area'])) * 100
merged['Percent_No_Train_Area'] = np.where(merged['Percent_No_Train_Area'] >= 99.9, 100, merged['Percent_No_Train_Area'])
merged['Percent_No_Bus_Area'] = np.where(merged['Percent_No_Bus_Area'] >= 99.9, 100, merged['Percent_No_Bus_Area'])
merged['Percent_SW_Bus_Area'] = (merged['SW_Bus_Length'] / merged['SW_Census_Total_Length']) * 100
merged['Percent_SW_Train_Area'] = (merged['SW_Train_Length'] / merged['SW_Census_Total_Length']) * 100
merged['Percent_SW_Grocery_Area'] = (merged['SW_Grocery_Length'] / merged['SW_Census_Total_Length']) * 100
merged['Ratio_SW_To_Street'] = (merged['SW_Census_Total_Length'] / merged['street_length'])
merged = merged.rename(columns={"PercentAreaNoGrocery":"Percentage_Grocery_Area"})
merged = merged.rename(columns={"Percent_No_Bus_Area":"Percentage_Bus_Area"})
merged = merged.rename(columns={"Percent_No_Train_Area":"Percentage_Train_Area"})
merged
```
